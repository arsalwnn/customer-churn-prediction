## Finding the best n_neighbor

````python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier


knn = KNeighborsClassifier()


param_grid = {'n_neighbors': range(1, 31)}


grid = GridSearchCV(knn, param_grid, cv=5, scoring='accuracy')
grid.fit(x_train, y_train)

print("Best k:", grid.best_params_)
print("Best accuracy:", grid.best_score_)

````

````python
#Knn
kn=KNeighborsClassifier(n_neighbors=3,metric="minkowski",p=2)
kn.fit(x_train,y_train)
````

````python
#Voting Classifier
vc = VotingClassifier(estimators=[('dt',dt),('lr',lr),('rf',rf),('knn',kn)])
vc.fit(x_train,y_train)
````

# 📌 SHAP (SHapley Additive exPlanations) Guide

This notebook shows how to use **SHAP** to interpret a machine learning model.  
SHAP values explain **how much each feature contributes** (positively or negatively) to a model’s prediction.

---



## 1️⃣ Importing and Creating the Explainer for DecisionTreeClassifier
```python


````python
import shap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

explainer= shap.Explainer(dt)
shap_values= explainer(x_test)
````

## 2️⃣ Checking Expected Value and SHAP Array

````python
print("explainer.expected_value:", explainer.expected_value)
print("shap_values.values.shape:", shap_values.values.shape)
print("shap_values.base_values.shape:", shap_values.base_values.shape)
````

## 3️⃣ Force Plot for a Single Prediction

````python
sample_idx = 0
class_idx = 0   # or 1 for the second class

shap.plots.force(
    explainer.expected_value[class_idx],          # base value for the selected class
    shap_values.values[sample_idx, :, class_idx], # SHAP values for this sample and this class
    x_test.iloc[sample_idx],                      # actual feature values for this sample
    matplotlib=True                               # render as a static matplotlib plot (not interactive JS)
)

````

## 4️⃣ Summary Plot (Feature Importance for a Class)

````python

shap.summary_plot(shap_values[:,:,0], x_test, plot_type="bar")

````

## 5️⃣ Numeric Feature Importance

````python
# Example for class 0
class_idx = 0
sv_class0 = shap_values[:,:,class_idx]

mean_abs_shap = np.abs(sv_class0.values).mean(axis=0)

feature_importance = pd.DataFrame({
    "Feature": x_test.columns,
    "Mean |SHAP value|": mean_abs_shap
}).sort_values(by="Mean |SHAP value|", ascending=False)

print(feature_importance)
````

### Evaluating Model Performance:

````python
models = {'Decision Tree':dt,  'Random Forest':rf,  'Logestic Regression':lr,  'Knn':kn,  'Voting Classifier':vc}
for name, model in models.items():
    y_pred = model.predict(x_test)
    accuracy= accuracy_score(y_test,y_pred)
    rmse= mean_squared_error(y_test,y_pred)
    print(f'{name} - Accuracy: {accuracy*100:.2f}%, RMSE: {rmse:.2f}')
````

### Normalize Numerical Festures using StandardScaler:

````python
scaler= StandardScaler()
df[df.columns]= scaler.fit_transform(df[df.columns])
df['Churn'] = df['Churn'].astype(int)
df['Churn'] = (df['Churn'] > 0.5).astype(int)
````

````python
x= df.drop('Churn',axis=1)
y= df['Churn']
````

### Convert Target to Categories:

### Split Dataset into Training and Testing Sets:

````python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
````

### Initialize and Train Models

````python
classifiers={
    'Decision Tree':DecisionTreeClassifier(),
    'Random Forest':RandomForestClassifier(),
    'Logistic Regression':LogisticRegression(),
    'Knn':KNeighborsClassifier()
}
voting_clf= VotingClassifier(estimators=[(name,clf) for name,clf in classifiers.items()])
classifiers['Voting Classifier']= voting_clf
````

````python
results= {}


for name,clf in classifiers.items():
    clf.fit(x_train,y_train)
    #training th classifier
    clf.fit(x_train,y_train)

    #making Predictions
    y_pred= clf.predict(x_test)

    #Calculating accuracy and RMSE
    accuracy= accuracy_score(y_test,y_pred)
    rmse= np.sqrt(mean_squared_error(y_test,y_pred))

    #Store the results
    results[name] = {'Accuracy':accuracy, 'RMSE':rmse}

results_df= pd.DataFrame(results).T
print(results_df)
````
